```python
import glob
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
import warnings
warnings.filterwarnings('ignore')
```

```python
df_organizations = pd.read_csv('organizations.csv')
df_investments = pd.read_csv('investments.csv')
df_funding_rounds = pd.read_csv('funding_rounds.csv')
df_investors = pd.read_csv('investors.csv', low_memory=False)
```

```python
df_invetment_investor = pd.merge(df_investments, df_investors, how='left', left_on='investor_uuid', right_on='uuid')
df_invetment_investor.columns
```

```python
df_invetment_investor['name_y'].isna().sum()
```

```python
df_invetment_investor_fr = pd.merge(df_invetment_investor, df_funding_rounds, how='left', left_on='funding_round_uuid', right_on='uuid')
df_invetment_investor_fr
```

```python
df_invetment_investor_fr.columns
```

```python
df_iif = df_invetment_investor_fr[['uuid_x', 'funding_round_uuid','investor_uuid','investor_name','investor_type','is_lead_investor','roles','country_code_x','region_x','city_x','investor_types','investment_count','total_funding','founded_on','closed_on','country_code_y','state_code_y','region_y','city_y','investment_type','announced_on','raised_amount','post_money_valuation','investor_count','org_uuid','org_name','lead_investor_uuids']]
df_iif
```

```python
df_iif.rename(columns={"uuid_x":"investment_uuid"}, inplace = True)
df_iif.rename(columns={"roles":"investor_roles"}, inplace = True)
df_iif.rename(columns={"investment_type":"Investment_round"}, inplace = True)

df_iif.rename(columns={"investor_types":"vc_types"}, inplace = True)
df_iif.rename(columns={"country_code_x":"Investor_country_code"}, inplace = True)

df_iif.rename(columns={"region_x":"investor_region"}, inplace = True)

df_iif.rename(columns={"city_x":"investor_city"}, inplace = True)

df_iif.rename(columns={"investor_types":"vc_types"}, inplace = True)
df_iif.rename(columns={"country_code_y":"org_country_code"}, inplace = True)

# df_iif.rename(columns={"state_code_y":"org_state_code"}, inplace = True)

df_iif.rename(columns={"region_y":"org_region"}, inplace = True)

df_iif.rename(columns={"city_y":"org_city"}, inplace = True)

df_iif
```

```python
df_iif[['investor_uuid','is_lead_investor', 'lead_investor_uuids']]
```

```python
df_iif['lead_investor_uuids'].fillna('none', inplace = True)
# lst = df_iif['lead_investor_uuids'].apply(lambda x: str(x.split(','))).to_numpy()
# np.unique(lst)
```

```python
# Determine the maximum number of comma-separated values
max_values = df_iif['lead_investor_uuids'].str.split(',').apply(len).max()

# Generate new column names
new_column_names = ['lead_investor_{}'.format(i) for i in range(1, max_values+1)]

# Split the column
df_iif[new_column_names] = df_iif['lead_investor_uuids'].str.split(',', expand=True)
```

```python
df_iif['lead_investor_available'] = False  # Initialize column with False
df_iif.loc[df_iif['lead_investor_uuids'] != 'none', 'lead_investor_available'] = True  # Set to True where condition is met
```
