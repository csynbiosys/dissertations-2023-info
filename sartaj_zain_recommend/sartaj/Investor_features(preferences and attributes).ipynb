```python
import glob
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
import warnings
warnings.filterwarnings('ignore')
```

```python
df_investments = pd.read_csv('investments.csv')
df_investors = pd.read_csv('investors.csv', low_memory=False)
df_iifo = pd.read_csv('investor_startup.csv')
df_organizations = pd.read_csv('organizations.csv')
investor_df_des= pd.read_csv('investor_text_embedding.csv')
```

```python
df_iifo.columns
```

```python
df_preference_of_investors = df_investments[['investor_uuid','funding_round_name']]
df_preference_of_investors['funding_round_name'] = df_preference_of_investors['funding_round_name'].apply(lambda x: x.split('-')[0])
df_preference_of_investors = df_preference_of_investors.pivot_table(index='investor_uuid', columns='funding_round_name', aggfunc=len, fill_value=0)

df_preference_of_investors
```

```python
df_preference_of_investors = df_preference_of_investors.div(df_preference_of_investors.sum(axis=1), axis=0)
df_preference_of_investors
```

```python
df_investors_feat = df_investors[['uuid', 'name', 'type', 'rank', 'roles',  'country_code',  'region',
       'city', 'investor_types', 'investment_count', 'total_funding_usd','founded_on', 'closed_on']]
```

```python

df_iifo['category_list'] = df_iifo['category_list'].str.split(',')
df_iifo['category_groups_list'] = df_iifo['category_groups_list'].str.split(',')

expanded_categories = df_iifo[['investor_uuid','category_list']].explode('category_list')
expanded_groups = df_iifo[['investor_uuid', 'category_groups_list']].explode('category_groups_list')

categories_encoded = expanded_categories['category_list'].str.get_dummies()
groups_encoded = expanded_groups['category_groups_list'].str.get_dummies()

categories_encoded['investor_uuid'] = expanded_categories['investor_uuid']
groups_encoded['investor_uuid'] = expanded_groups['investor_uuid']


category_investments = categories_encoded.groupby('investor_uuid').sum()
group_investments = groups_encoded.groupby('investor_uuid').sum()

category_investments

```

```python
category_investments = category_investments.div(category_investments.sum(axis=1), axis=0)
category_investments
```

```python
group_investments = group_investments.div(group_investments.sum(axis=1), axis=0)
group_investments
```

```python
df_categories = pd.merge(category_investments,group_investments, how='left', left_on='investor_uuid', right_on='investor_uuid')
df_categories
```

```python
df_categories.to_csv("investor_categories_preference.csv")
```

```python
df_country_preference = df_iifo[['investor_uuid','org_country_code']].pivot_table(index='investor_uuid', columns='org_country_code', aggfunc=len, fill_value=0)
df_country_preference = df_country_preference.div(df_country_preference.sum(axis=1), axis=0)
df_country_preference
```

```python
df_country_preference.columns
```

```python
df_region_preference = df_iifo[['investor_uuid','org_region']].pivot_table(index='investor_uuid', columns='org_region', aggfunc=len, fill_value=0)
df_region_preference = df_region_preference.div(df_region_preference.sum(axis=1), axis=0)
df_region_preference
```

```python
# df_city_preference = df_iifo[['investor_uuid','org_city']].pivot_table(index='investor_uuid', columns='org_city', aggfunc=len, fill_value=0)
# df_city_preference
```

```python
df_geo = pd.merge(df_country_preference,df_region_preference, how='left', left_on='investor_uuid', right_on='investor_uuid')
# df_geo = pd.merge(df_geo,df_city_preference, how='left', left_on='investor_uuid', right_on='investor_uuid')
df_geo
```

```python
df_geo.to_csv("investor_geo_preference.csv")
```

```python
df_investor_preferences = pd.merge(df_categories,df_geo, how='left', left_on='investor_uuid', right_on='investor_uuid')
df_investor_preferences
```

```python
df_investors_feat = pd.merge(df_investors_feat, df_organizations[['uuid','employee_count','category_list','category_groups_list']], how='left', left_on='uuid', right_on='uuid')
df_investors_feat['investment_count'].fillna(df_investors_feat['investment_count'].mean(), inplace=True)
df_investors_feat['investment_count'] = df_investors_feat['investment_count'].round()
investment_count = df_investors_feat['investment_count']
normalized_investment_count = (investment_count - investment_count.min()) / (investment_count.max() - investment_count.min())
df_investors_feat['investment_count'] = normalized_investment_count
df_investors_feat
```

```python
df_encoded = df_investors_feat['roles'].str.get_dummies(sep=',')
df_investor_roles = pd.concat([df_investors_feat['uuid'], df_encoded], axis=1)
df_investor_roles
```

```python
df_encoded = df_investors_feat['investor_types'].str.get_dummies(sep=',')
df_investor_types = pd.concat([df_investors_feat['uuid'], df_encoded], axis=1)
df_investor_types
```

```python
df_encoded = df_investors_feat['category_list'].str.get_dummies(sep=',')
df_category_list = pd.concat([df_investors_feat['uuid'], df_encoded], axis=1)
df_category_list
```

```python
df_encoded = df_investors_feat['category_groups_list'].str.get_dummies(sep=',')
df_category_groups_list = pd.concat([df_investors_feat['uuid'], df_encoded], axis=1)
df_category_groups_list
```
