```python
import glob
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
import warnings
warnings.filterwarnings('ignore')
```

```python
df_organizations = pd.read_csv('organizations.csv')
df_organization_descriptions = pd.read_csv('organization_descriptions.csv')
df_people_descriptions = pd.read_csv('people_descriptions.csv')
df_iifo = pd.read_csv('investor_startup_rel.csv')
```

```python
df_organization_descriptions
```

```python
df_organizations.columns
```

```python
investor_df_des = pd.merge(df_iifo[['investor_uuid','investor_name']],df_organization_descriptions[['uuid','description']], how='left', left_on='investor_uuid', right_on='uuid')
investor_df_des.drop('uuid',inplace = True,axis=1)
investor_df_des = pd.merge(investor_df_des,df_people_descriptions[['uuid','description']], how='left', left_on='investor_uuid', right_on='uuid')
investor_df_des.drop('uuid',inplace = True,axis=1)
investor_df_des = pd.merge(investor_df_des,df_organizations[['uuid','short_description']], how='left', left_on='investor_uuid', right_on='uuid')
investor_df_des.drop('uuid',inplace = True,axis=1)
```

```python
investor_df_des.drop_duplicates(inplace = True)
investor_df_des
```

```python
investor_df_des.isna().sum()
```

```python
investor_df_des['description'] = investor_df_des['description_x'].fillna(investor_df_des['description_y'])

investor_df_des['description'] = investor_df_des['description'].fillna(investor_df_des['short_description'])
investor_df_des['description'] = investor_df_des['description'].fillna(investor_df_des['investor_name'])
```

```python
investor_df_des = investor_df_des.drop(columns=['description_x','description_y','short_description','investor_name'], axis=1)

investor_df_des.isna().sum()
```

```python
investor_df_des.dropna(inplace=True)
```

```python
investor_df_des
```

```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python

embeddings = model.encode(investor_df_des['description'].tolist(), convert_to_tensor=True)


investor_df_des['embeddings'] = embeddings.tolist()
```

```python
embeddings_df = pd.DataFrame(embeddings.cpu())


investor_df_des = pd.concat([investor_df_des.reset_index(drop=True), embeddings_df.reset_index(drop=True)], axis=1)
```

```python
investor_df_des
```

```python
investor_df_des.to_csv('investor_text_embedding.csv')
```
