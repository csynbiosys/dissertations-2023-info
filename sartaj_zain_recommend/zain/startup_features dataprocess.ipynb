```python
import glob
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
import warnings
warnings.filterwarnings('ignore')
```

```python
df_iifo = pd.read_csv('investor_startup.csv')
```

```python
df_strt = df_iifo[['org_uuid','org_country_code', 'org_region', 'org_city','category_list', 'category_groups_list']]
df_strt.drop_duplicates(inplace=True)
df_strt
```

```python
df_encoded = df_strt['category_list'].str.get_dummies(sep=',')
df_category_list = pd.concat([df_strt['org_uuid'], df_encoded], axis=1)
df_category_list
```

```python
df_encoded = df_strt['category_groups_list'].str.get_dummies(sep=',')
df_category_groups_list = pd.concat([df_strt['org_uuid'], df_encoded], axis=1)
df_category_groups_list



```

```python
df_strt_country = df_strt[['org_uuid','org_country_code']].pivot_table(index='org_uuid', columns='org_country_code', aggfunc=len, fill_value=0)
df_strt_country


```

```python
df_strt_region = df_strt[['org_uuid','org_region']].pivot_table(index='org_uuid', columns='org_region', aggfunc=len, fill_value=0)
df_strt_region



```

```python
df_strt_attributes = pd.merge(df_category_list,df_category_groups_list, how='left', left_on='org_uuid', right_on='org_uuid')
```

```python
df_strt_attributes = pd.merge(df_strt_attributes,df_strt_country, how='left', left_on='org_uuid', right_on='org_uuid')
```

```python
df_strt_attributes = pd.merge(df_strt_attributes,df_strt_region, how='left', left_on='org_uuid', right_on='org_uuid')
```

```python
df_strt_attributes.to_csv('startup_features.csv')
```
