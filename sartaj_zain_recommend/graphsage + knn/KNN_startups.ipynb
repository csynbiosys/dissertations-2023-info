```python
import torch
from torch_geometric.data import Data
# from torch_geometric.transforms import KNNGraph
from torch_geometric.nn import knn_graph
from torch_geometric.nn import knn
import pandas as pd
```

```python
df_startup = pd.read_csv('train_for_knn.csv')
```

```python
df_startup
```

```python
org_features_df = df_startup.loc[df_startup['org_uuid'].drop_duplicates().index].drop(columns=['Unnamed: 0','org_uuid','description','embeddings'])
```

```python
org_features_df.fillna(0,inplace=True)
```

```python
org_features_df
```

```python
org_features_df.isna().sum()
```

```python
node_features = torch.tensor(org_features_df.values, dtype=torch.float32)
```

```python
# assert node_features.size() == (122426, 2229)
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python


node_features = node_features.to(device)
data = Data(x=node_features).to(device)


data.node_id = torch.arange(data.x.size(0))


# knn_graph_transform = KNNGraph(k=2, cosine = True, num_workers = 1)
# data = knn_graph_transform(data)

edge_index = knn_graph(data.x, k=5, cosine = True)

data.edge_index = edge_index

print(f"Edge index after constructing the k-NN graph:\n{data.edge_index}")

```
